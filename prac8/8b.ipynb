```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df=pd.read_csv('multiTimeline.csv',skiprows=1)
```

```python
df.columns=['Month','diet','gym','finance']
```

```python
df.Month=pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

```python
df.plot()
```

```python
df[['diet']].plot()
```

```python
diet=df[['diet']]
diet.rolling(12).mean().plot()
```

```python
gym=df[['gym']]
```

```python
gym.rolling(12).mean().plot()
```

```python
dm=pd.concat([diet.rolling(12).mean(),gym.rolling(12).mean()],axis=1)
dm.plot()
```

```python
diet.diff().plot()
```

```python
df.diff().plot()
```

```python
df.diff().corr()
pd.plotting.autocorrelation_plot(diet)
```
