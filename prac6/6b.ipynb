```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
data=dataset[['Annual_Income_(k$)','Spending_Score']].values
```

```python
d=[]
for i in range(1,11):
    model1=KMeans(n_clusters=i,random_state=56)
    model1.fit(data)
    d.append(np.mean(model1.inertia_))
```

```python
plt.plot(range(1,11),d)
```

```python
model2=KMeans(n_clusters=6)
```

```python
y=model2.fit_predict(data)
```

```python
plt.scatter(data[y == 0,0],data[y == 0,1],color="red")
plt.scatter(data[y == 1,0],data[y == 1,1],color="blue")
plt.scatter(data[y == 2,0],data[y == 2,1],color="green")
plt.scatter(data[y == 3,0],data[y == 3,1],color="black")
plt.scatter(data[y == 4,0],data[y == 4,1],color="yellow")
plt.scatter(data[y == 5,0],data[y == 5,1],color="pink")
```
